```python
import pandas as pd
import numpy as np
import re
import json
```

```python
filename = 'data_beautyofjoseon.json'
with open(filename, 'r') as f:
  data = json.load(f)
df = pd.DataFrame(data)
df = df.drop(['ig_username', 'name', 'title', 'bio', 'linktree'], axis=1)
```

```python
comments = []
i = 0
for item in df['posts']:
    for subitem in item.values():
        for element in subitem:
            if 'comment:' in element[0]:
                for comment in element[1]:
                    if comment not in comments and i % 2 == 0:
                        comments.append(comment)
                    i += 1
```

```python
comments.append('this is so ugly')
comments.append('gross')
comments.append('i hate you')
```

```python
new_df = pd.DataFrame(comments, columns=['comment'])
```
